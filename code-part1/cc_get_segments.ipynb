```python
file1 = open('test-100.wet','w')

for abc in resp.content.strip().split('\n'):
    if "/warc/" in abc:
        var = json.loads(abc)
        filename1 = var['filename']
        filename2 = filename1.replace('/warc/','/wet/')
        filename3 = filename2.replace('.warc.gz','.warc.wet.gz')
        print(filename3)
        file1.write(str(filename3)+"\n")
file1.close()
```
